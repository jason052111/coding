```python
import os
from PIL import Image

def TrainImg() :
    DATA_DIR = "C:/dog_identification/train"  # 資料夾路徑

    # 轉換ORIGTrainImg
    for filename in os.listdir(DATA_DIR):
        file_name = DATA_DIR + "/" + filename
        count = 1
        file_count = len(os.listdir(file_name))
        for breeds in os.listdir(file_name):
            if (breeds == ".DS_Store"):
                pass
            else:
                print("Loading: %s" % breeds)

                im = Image.open(file_name + "/" + breeds )

                width = 100
                height = 100
                reim = im.resize((width, height))
                saveName = "C:/dog_identification/resize_train/train/" + filename + "/" + str(count).zfill(len(str(file_count))) + ".jpg"
                reim.save(saveName)
                print(str(count) + ".jpg", "Resize and Save Complete")
                count = count + 1


def ValidImg() :
    DATA_DIR = "C:/dog_identification/valid"  # 資料夾路徑

    # 轉換ORIGTrainImg
    for filename in os.listdir(DATA_DIR):
        file_name = DATA_DIR + "/" + filename
        count = 1
        file_count = len(os.listdir(file_name))
        for breeds in os.listdir(file_name):
            if (breeds == ".DS_Store"):
                pass
            else:
                print("Loading: %s" % breeds)

                im = Image.open(file_name + "/" + breeds )

                width = 100
                height = 100
                reim = im.resize((width, height))
                saveName = "C:/dog_identification/resize_valid/valid/" + filename + "/" + str(count).zfill(len(str(file_count))) + ".jpg"
                reim.save(saveName)
                print(str(count) + ".jpg", "Resize and Save Complete")
                count = count + 1


def TestImg() :
    DATA_DIR = "C:/dog_identification/test"  # 資料夾路徑
    file_count = len(os.listdir(DATA_DIR))
    count = 1

    # 轉換ORIGTestImg
    for filename in os.listdir(DATA_DIR):
        if (filename == ".DS_Store"):
            pass
        else:
            print("Loading: %s" % filename)
            im = Image.open(DATA_DIR + "/" + filename)

            width = 100
            height = 100
            reim = im.resize((width, height))
            saveName = "C:/dog_identification/resize_test/" + filename
            reim.save(saveName)
            print(str(count) + ".jpg", "Resize and Save Complete")
            count = count + 1

```
