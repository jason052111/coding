```python
import os
import csv
from keras_preprocessing.image import img_to_array
from keras.utils import to_categorical
from PIL import Image
import numpy as np

np.random.seed(3)

label_dict = {0:"Airedale", 1:"Beagle", 2:"Bloodhound", 3:"Bluetick", 4:"Chihuahua", 5:"Collie",
              6:"Dingo", 7:"French Bulldog", 8:"German Sheperd",9:"Malinois", 10:"Newfoundland",
              11:"Pekinese", 12:"Pomeranian", 13:"Pug", 14:"Vizsla"}

reverse_label_dict = {v: k for k, v in label_dict.items()}

def preprcIMG_Valid() :
    train_img = []
    train_label = []
    valid_img = []
    valid_label = []

    DATA_DIR = "C:/dog_identification/"  # 圖片資料夾路徑
    with open('C:/dog_identification/dogs.csv', mode='r', encoding='utf-8') as f:
        reader = csv.reader(f)
        for row in reader:
            # images
            im = Image.open(DATA_DIR + "resize_" + row[2] + "/" + row[0])
            arr_img = img_to_array(im)
            if ( row[2] == "train" ) :
                train_img.append(arr_img)
            elif (row[2] == "valid" ) :
                valid_img.append(arr_img)

            print(row[0], 'has already in list!')

            # label
            if (row[2] == "train") :
                train_label.append(reverse_label_dict[row[1]])
            elif (row[2] == "valid") :
                valid_label.append(reverse_label_dict[row[1]])
            print(row[1], 'has already in list!')

    train_img = np.array(train_img)
    train_label = np.array(train_label)
    valid_img = np.array(valid_img)
    valid_label = np.array(valid_label)

    train_img_normalize = train_img.astype('float32') / 255.0
    valid_img_normalize = valid_img.astype('float32') / 255.0

    train_label_OneHot = to_categorical(train_label)
    valid_label_OneHot = to_categorical(valid_label)
    return train_img, valid_img, train_label, valid_label, train_img_normalize, valid_img_normalize, train_label_OneHot, valid_label_OneHot

def preprcTest() :
    test_img = []
    test_label = []

    DATA_DIR = "C:/dog_identification/resize_test/"  # 圖片資料夾路徑
    # 轉換ORIGTestImg
    for filename in os.listdir(DATA_DIR):
        if (filename == "test_data.xlsx"):
            pass
        else:
            print("Loading: %s" % filename)
            im = Image.open(DATA_DIR + filename)
            arr_img = img_to_array(im)
            test_img.append(arr_img)
            print(filename, 'has already in list!')
            test_label.append(0)
            print(filename, 'has already in list!')
    test_img = np.array(test_img)
    test_label = np.array(test_label)
    test_img_normalize = test_img.astype('float32') / 255.0
    test_label_OneHot = to_categorical(test_label)
    return test_img, test_label, test_img_normalize, test_label_OneHot

```
